# Data Cleaning

```python
data_1 = data_1.drop(['ID','Start time','Completion time','Email','Name', 'Last modified time'], axis = 1)
```

```python
data = data_1.rename(columns = {'In which academic program are you currently enrolled?\n' : 'Program',
                           'What is the Month and Year of your intake?' : 'Enrolled_time',
                           'What is your country of origin?\n': 'Country',
                           'Could you provide state/province of origin for above mentioned country?' : 'Province',
                           'Which of the following age group best describes you?\n' : 'Age',
                           'Which gender do you identify yourself as?\n' : 'Gender',
                           "As a student in the School of IT, what do you believe are the essential skills and tools you need for the program you're enrolled in?\n " : 'Basic_skills',
                           'What was the highest level of education you completed (prior to starting this program)?\n' : 'Previous_education',
                           'What was your major filed of study as per your highest degree or level of education?\n' : 'Majors',
                           'Do you have any prior work experience?': 'Experience',
                           'How many years of work experience do you have?' : 'year_experience',
                           'In which sector/industry have you had previous work experience ?': 'Sector_worked',
                           'Please provide the zip code for your current accommodation in Windsor.' : 'Current_zipcode',
                           'What type of accommodation do you have?' : 'accomodation',
                           'How much monthly rent do you pay including utilities?' : 'rent_payed',
                           'Ease of finding accommodation' : 'accommodation_rate',
                           'Quality of accommodation' : 'accommodation_quality',
                           'Affordability of accommodation' : 'accommodation_affordability',
                           'What is your current employment status?':'Employment_status',
                           'What is your average monthly income?' : 'income_avg',
                           'Are you aware of or use any of the student services offered by the college/SRC?' : 'service_awareness',
                           'Which services do you use?' : 'service_usage',
                           'Please provide any additional feedback, appreciation, or suggestions regarding your experience.' : 'Feedback'})
```

```python
data = data.dropna(subset='Program')
```

```python
columns = ['service_usage', 'year_experience', 'Sector_worked','Feedback']

for column in columns:
    data[column] = data[column].fillna('None')
```

```python
data['Current_zipcode'] = data['Current_zipcode'].str.replace(r'\W', '').str.upper()
```

```python
data['Enrolled_time'] = pd.to_datetime(data['Enrolled_time'])

# Create new columns for the year and month
data['Enrolled_year'] = data['Enrolled_time'].dt.year
data['Enrolled_month'] = data['Enrolled_time'].dt.month

# Drop the original 'Enrolled_time' column
data = data.drop(columns=['Enrolled_time'])
```

```python
data['Sector_worked']  = data['Sector_worked'].replace({'Back End Developer' : 'IT', 'IT Project Management, Robotics Product Development, Social Media Marketing, SEO, Sales, Leadership Role.' : 'IT',
                                                       'IT software develoment':'IT','python web developping':'IT','Software developer':'IT',
                                                       'Software Development':'IT','Web Development':'IT','In IT as a Technical Support Engineer, Network Associate Engineer, and Network Engineer':'IT',
                                                       'Institute - as a Computer Operator and Trainer':'IT','it' : 'IT','IT as Data Analyst':'IT','IT Industry':'IT',
                                                       'IT industry':'IT','IT Industry - Data Analyst':'IT','Software Quality Analyst, Software Testing' :'IT',
                                                       'Data Analyst':'IT','Data analytics':'IT','Data Analytics for eCommerce Sector':'IT','Software Engineer':'IT','Computer' :'IT',
                                                       'Telecom' : 'IT','ITES' : 'IT','Information Technology' : 'IT','IT ' : 'IT','semi tech' : 'IT'})

data['Sector_worked']  = data['Sector_worked'].replace({'Banking and Finance':'Finance','Finance, Accounting, Procurement, Auditing' :'Finance',
                                                       'Finance, Sales.':'Finance','Financial Services':'Finance','Insurance Domain Sunlife Financial ':'Finance',
                                                       'Fintech, CPA' : 'Finance','Finance ' : 'Finance'})

data['Sector_worked']  = data['Sector_worked'].replace({'Oil & gas' : 'Energy','Oil and gas' :'Energy','Oil and gas ' : 'Energy','Energy Sector' : 'Energy'})

data['Sector_worked']  = data['Sector_worked'].replace({'Education (Assistant Professor)' : 'Education','Bookkeeping' : 'Education','Teaching Primary level' : 'Education'})

data['Sector_worked']  = data['Sector_worked'].replace({'Food Industry':'Food'})

data['Sector_worked']  = data['Sector_worked'].replace({'Service work and private contractor' :'Construction','Construction and project management ' : 'Construction','construction' : 'Construction',
                                                        'Consturction' : 'Construction'})

data['Sector_worked']  = data['Sector_worked'].replace({'marketing consulting, IT tech solution & CRM(customer relationship management) ' :'Marketing','Marketing and Business sector' : 'Marketing',
                                                        'Digital Marketing' :'Marketing'})

data['Sector_worked']  = data['Sector_worked'].replace({'security survilence and car garage' : 'General Labour','Roofing , Moving , Typist' : 'General Labour','General Labor, manufacturing' : 'General Labour',
                                                        'Manufacturing ' : 'General Labour','As a Assistant Computer Hardware Engineer, Customer Service Representative, Cashier' : 'General Labour'})

data['Sector_worked']  = data['Sector_worked'].replace({'Transportation ' : 'Logistics'})

data['Sector_worked']  = data['Sector_worked'].replace({'Business Development(IT, Pharmaceuticals and Electricity)':'Business','Business ' :'Business'})
```

```python
data['rent_payed']  = data['rent_payed'].replace({"$0- parent's house":'0','$w0' :'0','i live at home' : '0','I dont pay' : '0'})
data['rent_payed']  = data['rent_payed'].replace({'700' : '$600-$1000','630' : '$600-$1000','650' :'$600-$1000','695' : '$600-$1000',
                                                 '620' : '$600-$1000','850' :'$600-$1000','630$' :'$600-$1000','750' :'$600-$1000','675' :'$600-$1000'})
data['rent_payed']  = data['rent_payed'].replace({'1300' : '$1000+','1350':'$1000+','$2500':'$1000+','$1400' :'$1000+','1100':'$1000',
                                                 '~1000':'$600-$1000','1200' : '$1000+','$1600' : '$1000+','$1350' : '$1000+','1700':'$1000+',
                                                '1000+' : '$1000+','1050':'$1000+','1000':'$600-$1000','$1000' : '$600-$1000','600':'$600-$1000','$200':'less than $300'})
```

# Univariate Analysis

```python
data['Age'] = pd.Categorical(data['Age'], categories=['17-25', '26-35', '36-45'], ordered=True)

plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Age', data=data, palette='viridis')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
likert_columns = ['accommodation_rate', 'accommodation_quality', 'accommodation_affordability',
                  'Reliable commute', 'Availability of daily needs']

# Convert Likert scale columns to numeric
data[likert_columns] = data[likert_columns].apply(pd.to_numeric, errors='coerce')

# Summary statistics
summary_stats = data[likert_columns].describe()

# Plotting bar graphs for each Likert scale column
for column in likert_columns:
    counts = data[column].value_counts().sort_index()
    percentages = counts / counts.sum() * 100

    plt.figure(figsize=(8, 6))
    plt.bar(counts.index, percentages)
    plt.xlabel(column)
    plt.ylabel('Percentage')
    plt.title('Likert Scale: {}'.format(column))
    plt.xticks(counts.index)
    plt.show()
```
